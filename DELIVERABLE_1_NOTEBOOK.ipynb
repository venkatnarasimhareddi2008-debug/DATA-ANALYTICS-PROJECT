```python
import pandas as pd

df = pd.read_csv('Dataset.csv')
df.head()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.upper()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'PURCHASE_AMOUNT_(USD)':'PURCHASE_AMOUNT'})
```

```python
labels = ['Young', 'Adult', 'Mid-age', 'Old']
df['AGE_GROUP'] = pd.qcut(df['AGE'], q=4, labels=labels)
```

```python
df[['AGE','AGE_GROUP']].head(11)
```

```python
frequency_mapping = {
    'Weekly': 7,
    'Fortnightly': 14,
    'Monthly': 30,
    'Quarterly': 90,
    'Every 3 Months': 90,
    'Bi-Weekly': 14,
    'Annually': 365
}
df['PURCHASE_FREQUENCY'] = df['FREQUENCY_OF_PURCHASES'].map(frequency_mapping)
```

```python
df[['AGE', 'AGE_GROUP', 'FREQUENCY_OF_PURCHASES', 'PURCHASE_FREQUENCY']].head(11)
```

```python
(df['DISCOUNT_APPLIED'] == df['PROMO_CODE_USED']).all()
```

```python
df = df.drop('PROMO_CODE_USED', axis=1)
```

```python
import numpy as np
```

```python
df['Estimated_Total_Spend'] = df['PURCHASE_AMOUNT'] * (df['PREVIOUS_PURCHASES'] + 1)
```

```python
df[['AGE', 'AGE_GROUP', 'FREQUENCY_OF_PURCHASES', 'PURCHASE_FREQUENCY','Estimated_Total_Spend']].head(11)
```

```python
binary_cols = ['SUBSCRIPTION_STATUS', 'DISCOUNT_APPLIED',]
for col in binary_cols:
    df[col] = df[col].apply(lambda x: 1 if str(x).strip().lower() == 'yes' else 0)
```

```python
df['Churn_Risk'] = np.where((df['REVIEW_RATING'] < 3.0) & (df['SUBSCRIPTION_STATUS'] == 0), 1, 0)
```
